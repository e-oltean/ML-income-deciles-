```python
# import pandas and numpy modules
import pandas as pd
import numpy as np
```

```python
#import the data
data=pd.read_csv('/content/threshold-income-for-each-decile-after-tax-lis.uk.csv', header=None)
```

```python
data
```

```python
#transpose the data
transposed_data = data.transpose()
```

```python
transposed_data
```

```python
# eliminate the header
transposed_data1=transposed_data[3:]
```

```python
transposed_data1
```

```python
#adding the population income expressed in deciles

df = transposed_data1.assign(deciles=[ 10, 20, 30, 40, 50, 60, 70, 80, 90])
```

```python
df
```

```python
# selecting the the data. X represent the the UK population's income divided into deciles for the time interval from 1968 and 2022 .
# y represent the deciles (percentage) of population that has an income or lower than the respectiv percentage (10%....90%)
X=df.iloc[:, 1:55].values
y=df.iloc[:, -1:].values
```

```python
# split the data (X and y) into training and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=1)
```

```python

# traing the model (regression)
from sklearn.linear_model import LinearRegression
regressor=LinearRegression ()
regressor.fit(X_train, y_train)
```

```python
# predicting the the income values based on the trained model applied to test data
y_pred=regressor.predict(X_test)
y_pred
```

```python
# calculating the coefficient of determination and importing the module
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
r2
```
